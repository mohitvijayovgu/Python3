```python
from pydantic import BaseModel
```

```python
class employee_schema(BaseModel):
    name: str
    id: int
    department: str
```

```python
obj = employee_schema(name="John Doe", id=123, department="HR")
print(obj)
```

```python
obj2 = employee_schema(**{"name": "John Doe", "id": 123, "department": "HR"})
print(obj2)
```

```python
from pydantic import StrictInt
class employee_schema(BaseModel):
    name: str
    id: StrictInt
    department: str
obj4 = employee_schema(name="John Doe", id=123, department="HR")
print(obj4)
```
