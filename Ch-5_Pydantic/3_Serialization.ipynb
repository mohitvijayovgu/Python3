```python
from pydantic import BaseModel, Field, EmailStr
from typing import Optional, Literal

class employee_schema(BaseModel):

    name: str = Field(..., min_length=3, max_length=50, description="The employee's full name")
    age: int = Field(..., gt=18, lt=65, description="The employee's age")
    email: EmailStr = Field(..., min_length=5, max_length=100, description="The employee's email address")
    department:Optional[str] = None
    gender: Literal['Male', 'Female', 'Other']
```

```python
obj = employee_schema(name="John Doe", age=30, email="john.doe@example.com", department="HR", gender="Male")
print(obj)
```
